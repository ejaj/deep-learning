# Beer's Law
It is a law that models the physics of x-ray attenuation. The following derivation has been inspired by *[1]*.

<div>
<img src="./img_01_beer_law.png" width="500"/>
</div>


* let a x-ray beam have Intensity $I(x)=E\cdot N(x)$ with $E$ being the energy of the $N(x)$ photons passing through $x$ each second
* let the beam traverse a slab of material of thickness $\Delta x$, with an attenuation coefficient $A(x)=k$ inside the slab and zero elsewhere.
* the proportional loss of photons that are blocked by the slab with thickness $\Delta x$ is then $p(\Delta x)=-A(x)\Delta x$.

By multiplying both sides by entrying intensity $I(x)$ we obtain

\begin{align}
\underbrace{I(x)p(\Delta x)}_{\Delta I} = -A(x)\Delta x I(x)
\end{align}


if we now consider an inifinitesimal slab for $\Delta x\to0$, then we obtain *Beer's law*

\begin{align}
\frac{dI}{dx}(x)=-A(x)I(x)
\end{align}

We can now solve this differential equation. First we rearrange it (and we drop the extra "$x$")
\begin{align}
\frac{dI}{I}=-A(x)dx
\end{align}

then we integrate it within the slab between $x$ and $x+\Delta x$

\begin{align}
\int_x^{x+\Delta x}\frac{dI(y)}{I(y)} dy=- \int_x^{x+\Delta x} A(y)dy
\end{align}

note that the left hand side is a known integral (see *[2]*) and in particular by making the sostututions $I(y)=t$ and $dI(y)dy=I'(y)dy=dt$

\begin{align}
\int \frac{dt}{t} = log|t|+c
\end{align}

hence we obtain
\begin{align}
log[I(y)]|^{y=x+\Delta x}_{y=x} = log[I(x+\Delta x)] - log[I(x)]=- \int_x^{x+\Delta x} A(y)dy
\end{align}

if we finally remember that $A(x)=k$ between $x$ and $x+\Delta x$, then the integral on right hand side simply becomes $k\Delta x$, to finally give 

\begin{align}
log[I(x+\Delta x)] - log[I(x)]=- k\Delta x
\end{align}

which can be rearranged as
\begin{align}
I(x+\Delta x) = I(x)e^{-k\Delta x}
\end{align}

which by simplyfing notation as incident beam $I_0$ and outgoing beam $I_1$, while calling the thickness $x$ instead of $\Delta x$, may be more often reported as
\begin{align}
I_1 = I_0e^{-kx}.
\end{align}

In general, however, the attenuation is more complex and depends on attenuation of all the material encountered along the propagation of the x-ray beam until it hits the detector. Given a 2D slice of material with attenuation represented by an unknown function $f(x,y)$, whre $x$ and $y$ are the locations - or pixels - then

\begin{align}
I = I_0e^{-\int f(x,y) ds}
\end{align}

where the integration in $ds$ symbolizes the integration along a line.

**The goal of tomographic reconstruction is to calculate $f(x,y)$.**
